```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
dataset = concrete_data.values
```

```python
#normalising the data
dataset_norm = []
for i in range(9):
    _mean = np.mean(dataset[:,i])
    _std = np.std(dataset[:,i])
    _norm = (dataset[:,i] - _mean)/_std
    dataset_norm.append(_norm)
```

```python
dataset_normalised = np.asarray(dataset_norm).T
```

```python
# split the concrete_data set into predictors (inputs) and target (output)
X = dataset_normalised[:,0:7]
y = dataset_normalised[:,8]
```

```python
#One hidden layer of 10 nodes, and a ReLU activation function
#Use the adam optimizer and the mean squared error as the loss function.
# Create the neural network in a function so we can use it multiple times in the
# subsequent sections
def baseline_model():
    #create the model
    model = Sequential()
    model.add(Dense(7, input_dim=7, init='normal', activation='relu'))
    model.add(Dense(10,activation = 'relu'))
    model.add(Dense(1))
    #compile the model
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return model
model = baseline_model()
```

```python
#Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_split helper function from Scikit-learn.
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 72)
```

```python
X_train.shape
```

```python
mse = []
for i in range(50):
    model = baseline_model()
    model.fit(X_train,y_train, epochs = 50)
    y_pred = model.predict(X_test)
    mse.append(mean_squared_error(y_test,y_pred))
```

```python
mean = np.mean(mse)
mean
```

```python
std = np.std(mse)
std
```

```python

print("Mean squared errors for 50 regression models: mean = %.3f, std dev = %.3f" %(mean, std))
```
